```python
import numpy as np
import pandas as pd
import os

os.chdir(r"C:\\Users\\USER\\Desktop\\Project\\Proceed\\Seoul_population_movement")
```

### 1. 가장 많이 출발한 지역은 어디일까?

```python
### 1. 가장 많이 출발한 지역은 어디일까?
departure_count = sep_movement['출발지'].value_counts()

# 데이터 프레임화
departure_count = pd.DataFrame(departure_count)
departure_count  # 2_2
```

### 2. 가장 많이 도착한 지역은 어디일까?

```python
### 2. 가장 많이 도착한 지역은 어디일까?
arrive_count = sep_movement['도착지'].value_counts()

# 데이터 프레임화
arrive_count = pd.DataFrame(arrive_count)
arrive_count # 2_3
```

### ※ 성별, 연령별 따라 출발지와 도착지에 변화가 있을까?

1. 성별에 따라 변화가 있는지 확인해보자

```python
M_sep_movement = sep_movement[sep_movement['성별'] == 'M']
```

```python
M_departure_count = M_sep_movement['출발지'].value_counts()
M_departure_count = pd.DataFrame(M_departure_count)

M_departure_count # 2_4
```

```python
F_sep_movement = sep_movement[sep_movement['성별'] == 'F']
```

```python
F_departure_count = F_sep_movement['출발지'].value_counts()
F_departure_count = pd.DataFrame(F_departure_count)

F_departure_count # 2_5
```

도착지도 확인해본 결과

#### 결론 :  1. 성별에 따른 출발지 도착지 변화는 크게 없었다.  

2. 연령별에 따라 변화가 있을까?

연령대는 10대, 20-30대, 40-50대, 60대 이상으로 나눴다.


```python
## 2-1  10대에 따른 출발, 도착지 변화
age_10 = sep_movement[sep_movement['나이'] <= 15]

age_10_count = age_10['출발지'].value_counts()
age_10_departure = pd.DataFrame(age_10_count)

age_10_departure # 2_6
```

```python
## 2-2 2030대 연령별
age_2030 = sep_movement[(sep_movement['나이'] <= 35) & (sep_movement['나이'] >= 20)]

age_2030_count = age_2030['출발지'].value_counts()
age_2030_departure = pd.DataFrame(age_2030_count)

age_2030_departure  # 2_7
```

```python
## 2-3 4050대 연령별
age_4050 = sep_movement[(sep_movement['나이'] <= 55) & (sep_movement['나이'] >= 40)]

age_4050_count = age_4050['출발지'].value_counts()
age_4050_departure = pd.DataFrame(age_4050_count)

age_4050_departure   # 2_8
```

```python
## 2-4 60대 이상 연령별
age_60 = sep_movement[sep_movement['나이'] >= 60]

age_60_count = age_60['출발지'].value_counts()
age_60_departure = pd.DataFrame(age_60_count)

age_60_departure  # 2_9
```

각 연령대별 출발지와 도착지를 모두 확인한 결과

#### 결론 2. 연령대별 출발지와 도착지의 차이가 크지 않았다.
#### 또한 강남구, 서초구, 송파구에 사람들이 많이 방문한다.

※ 추가적으로 경기도, 인천은 시와 구로 구별하였으나 다른 도나 시의 묶여서 계산되었습니다.

###  3. 출퇴근 시간 대 이동 인구가 많은 지역은 어디일까?

출퇴근 시간대인 7시-8시, 18-19시에 이동 인구가 많은 지역이 어디인지 알아보자

```python
commute_time = [7,8,18,19]
```

```python
commute_movement = sep_movement[sep_movement['도착한 시간'].isin(commute_time) ]
```

```python
commute_movement
```

```python
commute_arrive_count = commute_movement['출발지'].value_counts()

df_commute_arrive = pd.DataFrame(commute_arrive_count)

df_commute_arrive.head(10)
```

```python
commute_departure_count = commute_movement['도착지'].value_counts()

df_commute_departure = pd.DataFrame(commute_departure_count)

df_commute_departure.head(10)
```

#### 3.1 요약

출퇴근 시간대 이동 인구가 많은 지역으로 서울시 강남구, 서초구, 송파구, 영등포구, 강서구로 나타났다.

###  4. 출퇴근 시간 대를 제외한 이동 인구가 많은 지역은 어디일까?

```python
daily_movement = sep_movement[~sep_movement['도착한 시간'].isin(commute_time) ]
```

```python
daily_arrive_count = daily_movement['출발지'].value_counts()

df_daily_arrive = pd.DataFrame(daily_arrive_count)

df_daily_arrive.head(10)
```

```python
daily_departure_count = daily_movement['도착지'].value_counts()

df_daily_departure = pd.DataFrame(daily_departure_count)

df_daily_departure.head(10)
```

#### 4.1 요약

출퇴근 시간대를 제외한 시간 대의 이동 인구가 많은 지역으로 서울시 강남구, 서초구, 송파구, 영등포구, 강서구로 나타났다.

출퇴근 시간대가 이동 지역에 큰 영향을 미치지 않는다.
